# **Keskiarvoja**

```python
# lasketaan kuukausittaiset keskiarvot päätöshinnoille: (M=kuukausi, Q=neljännesvuosi, Y=vuosi)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(14, 5))
nordea['Close'].resample('M').mean().plot(ax=ax[0])
aktia_bank['Close'].resample('M').mean().plot(ax=ax[1])
```

```python
# lasketaan vaihdon kokonaismäärät kuukausitasolla:
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(14, 5))
nordea['Volume'].resample('M').sum().plot(ax=ax[0])
aktia_bank['Volume'].resample('M').sum().plot(ax=ax[1])
```

Kuukausittaiset päätöhinnat tukevat aikaisempia johtopäätkösiä. Mutta vaihdon kuukausittaiset keskiarvot ovat mielestäni mielenkiintoisia. Nordean osakkeella on monta piikkiä, mistä yksi on korona-ajan alun romahdukseen liittyvä. Tämän jälkeen isoja määriä osakkeita kuukausittain on liikkunut melko tasaisesti. Mikä on varmasti seurausta jatkuvista positiivisista uuttisista Nordean tuloskunnosta sekä korkojen noususta.

Aktialla taas tuo kuukuasittainen vaihto on ollut paljon maltillisempaa ja tasaista, mutta sielläkin on selkeät piikit. Korona-ajan alun romahdus on selkeänä piikkinä. Toinen piikki löytyy loppuvuodesta 2021, mutta en osaa arvioida, mistä tämä johtui. Niin kuin sanoin, en ole seurannut Aktiaa tarkemmin.

# **Liukuvia keskiarvoja**

```python
# Piirretään samaan kuvaan päivän päätös (Close) liukuvia keskiarvoja Nordealta:
# Arvot sellaisenaan:
nordea['Close'].plot(figsize=(13,6))
# liukuva keskiarvo: (50 arvoa)
nordea['Close'].rolling(50).mean().plot()
# liukuva keskiarvo: (200 arvoa)
nordea['Close'].rolling(200).mean().plot()
# lisätään selitelaatikko:
plt.legend(['Arvot sellaisenaan', '50 päivän liukuva keskiarvo', '200 päivän liukuva keskiarvo'])
```

```python
# Piirretään samaan kuvaan päivän päätös (Close) liukuvia keskiarvoja Aktialta:
# Arvot sellaisenaan:
aktia_bank['Close'].plot(figsize=(13,6))
# liukuva keskiarvo: (50 arvoa)
aktia_bank['Close'].rolling(50).mean().plot()
# liukuva keskiarvo: (200 arvoa)
aktia_bank['Close'].rolling(200).mean().plot()
# lisätään selitelaatikko:
plt.legend(['Arvot sellaisenaan', '50 päivän liukuva keskiarvo', '200 päivän liukuva keskiarvo'])
```

50 ja 200 päivän liukuvat keskiarvot päivän päätös kurssista antavat hyvän kuvan siitä, milloin on hyvä ostaa sekä myydä osaketta. Nordean ja Aktian kursseja tarkastellessa nämä hyvät oston paikat ovat myös näin todettavissa.


# **Muutosprosentit**

```python
# Lasketaan prosentuaaliset muutokset, tarkastellaan yhä päätöshintoja:
nordea['nordea_muutos%'] = nordea['Close'].pct_change()
aktia_bank['aktia_bank_muutos%'] = aktia_bank['Close'].pct_change()

# Molemmilta puuttuu ensimmäinen arvo puuttuu
nordea

```

```python
# Muodostetaan uusi dataframe, siihen listataan molempien osakkeiden muutosprosentit:
muutokset = pd.concat([nordea['nordea_muutos%'],aktia_bank['aktia_bank_muutos%']], axis=1)
# korvataan puuttuvat arvot nollilla:
muutokset = muutokset.fillna(0)
muutokset*100
```

```python
# Tarkastellaan pahinta korona-aikaa, aikaväliltä 1.12.2019-31.12.2020 ja siirrytään desimaaliluvuista prosentteihin
(muutokset['2019-12-1':'2020-12-31']*100).plot(figsize=(13,6))
# y-akselin otsikko
plt.ylabel('Muutosprosentti')
# merkitään 0-kohta paremmin
plt.axhline(color='#262626')
```

Pahimman korona-ajan (omamääritelmä 1.12.2019-31.12.2020) muutosprosenttia tarkasteltaessa, voi todeta, että isoimmat heilahtelut tapahtuivat maaliskuussa 2020. Tämä oli kuukausi, milloin lähes kaikkea aloitettiin rajoittaa ensimmäisen kerran. Niin Nordea, kuin Aktia ovat ottaneet osumaa silloin. Molempien synkin päivä on leikannut osakkeen hinnasta yli 10 prosenttia. Korona-ajan pohjalta tapahtunut nousu ei ole ollut yhtä rajua päiväkohtaisesti, vaan nousu on jakautunt hieman pidemmälle aikavälille.

```python
# Tarkastellaan Venäjän hyökkäyssodan (24.2.2022) aloitusaikaa , aikaväliltä 23.2.2022-31.3.2022 ja siirrytään desimaaliluvuista prosentteihin
(muutokset['2022-2-23':'2022-3-31']*100).plot(figsize=(13,6))
# y-akselin otsikko
plt.ylabel('Muutosprosentti')
# merkitään 0-kohta paremmin
plt.axhline(color='#262626')
```

Venäjän aloittaman hyökkäyksen johdosta Nordean osakkeen arvosta suli noin 10 prosenttia helmi-maaliskuussa 2022. Sieltä pohjalta Nordea nousi yli 14 prosenttia. Joten oikea-aikainen osto olisi tuonut muhkean tuoton.

Aktian liikehdintä on ollut paljon maltillisempaa samaisena aikana. Käyrä mukailee kyllä Nordean vastaavaa, mutta paljon pienempinä liikkeinä. Aktiasta voisi sanoa, että on se on hieman pienempi riskinen osake.

```python
# tunnuslukuja muutosprosenteista:
(muutokset*100).describe().round(2)
```

```python
# sama graafisesti:
sns.boxplot(data=muutokset*100)
```

```python
# poimitaan isot muutokset ja muutetaan prosenttimuotoon:
muutokset[(abs(muutokset['nordea_muutos%'])>0.1) | (abs(muutokset['aktia_bank_muutos%'])>0.1)]*100
```

Taulukko kertoo ja tukee aiempaa, eli korona-aika on ollut hyvin vaikea ja vaikeasti ennustettava. Eikä kukaan ole oikein tiennyt,mitä tapahtuu. Tämä näkyy osakkeiden suurissa heilahteluissa.

```python
# muutosten korrelaatiot:
muutokset.corr()
```

Korrelaatiot ovat positiivisia, mikä tukee sitä, että Nordean ja Aktian muutoksissa on samanlaisuutta. Tämä tukee silmällä todettua huomiota, mitä kuvioista olen todennut.

```python
# Tarkastellaan yhteyttä myös hajontakuviolla:
plt.scatter(muutokset['nordea_muutos%'],muutokset['aktia_bank_muutos%'])
plt.xlabel('Nordean muutosprosentti desimaalilukuna')
plt.ylabel('Aktian muutosprosentti desimaalilukuna')
```

Hajontakuvion täplät kertovat, että osakkeet ovat reagoineet usein saman suuntaisesti eli kun toinen on noussut, niin toinenkin on noussut. Samanlailla laskut ovat tapahtunet myös. Hajontakuvio tukee aiempia havaintoja tarkastelujakson datasta.

```python
# lasketaan liukuva korrelaatio:
muutokset['nordea_muutos%'].rolling(100).corr(muutokset['aktia_bank_muutos%']).plot()
```

Liukuva korrelaatio on ollut Nordean ja Aktian välillä positiivinen koko tarkasteltavan ajanjakson ajan. Mikä tukee aiempia toteamuksia datan analysoinnista. Lisäksi, se, että molemmat yritykset ovat rahoitusalalta, selittää tätä positiivista korrelaatiota

```python
# lasketaan 200 päivän liukuva volatiliteetti
plt.figure(figsize=(13,7))
(muutokset['nordea_muutos%'].rolling(200).std()*(252**0.5)).plot(label='Nordea', legend=True)
(muutokset['aktia_bank_muutos%'].rolling(200).std()*(252**0.5)).plot(label='Aktia', legend=True)
```

Kuvaaja kertoo sen, mitä aikaisemmin tässä notebookissa olen todennut. Korona-aika on ollut suuri riskistä aikaa sekä Venäjän hyökkäylsen ensimmäinen vuosi. Nordean osake on ollut isomman vaihtelun vallassa. Tämä volatiliteetti-kaavio myös kertoo sen ja sen, että näistä kahdesta osakkeesta, tämä on se isompi riskinen. Näistä kahdesta rahoitusalan osakkeesta Aktia on hieman matalariskisempi ja sopii semmoiselle sijoittajalle, kuka ei halua ottaa isoja riskejä.
